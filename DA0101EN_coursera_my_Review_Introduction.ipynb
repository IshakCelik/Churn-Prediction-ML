#chi-square

````python
df.info()
````

````python
print(df.aspiration.value_counts())
print('---')
df['fuel-type'].value_counts()
````

````python
cont_table=pd.crosstab(df["aspiration"],df["fuel-type"]) #, normalize='index'
cont_table
````

````python
plt.figure(figsize=(12,12))
sns.heatmap(cont_table, annot=True, cmap="YlGnBu")
````

````python
# Chi-square test of independence.
c, p, dof, expected = stats.chi2_contingency(cont_table)
# Print the p-value
print('c=', c.round(3), 'p=', p, 'dof=', dof, 'expected=', expected.round(3))
````

````python
result=stats.chi2_contingency(cont_table, correction=True)
print(result, '/n')
print('p_value=', result[1])
````
